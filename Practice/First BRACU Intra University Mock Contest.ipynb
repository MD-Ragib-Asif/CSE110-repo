So they asked you to play the referee role. Can you help them out?

Given two arrays of 7 lengths where the first one is Umair’s and the second one is Zuaina’s expense on food in a week can you tell who is the winner?

Input
The first line of input consists of 7 integers which represent the cost of Umair’s expense on a week. The second line of input consists of 7 integers which represent the cost of Zuaina's expense on a week. It is guaranteed that their total expense in a week won't be the same. so there won't be any draw case.

Output
If Umair wins print “umair wins the buffet” (without quotes). If Zuaina wins then print “zuaina wins the buffet” (without quotes). Check the samples to understand better

Samples
Input	Output
12 25 100 2 63 3 1 
20 12 80 2 50 3 1
umair wins the buffet
Input	Output
20 12 80 2 50 3 1
12 25 100 2 63 3 1 
zuaina wins the buffet


```python
def array():
    list_string=input()
    list1=[int(x) for x in list_string.split()]
        
    return list1

umair=array()
zuaina=array()

def sumation(list_check):
    total=0
    for i in range(7):
        total+=list_check[i]
    
    return total

umair_total=sumation(umair)
zuaina_total=sumation(zuaina)

if umair_total>zuaina_total:
    print('umair wins the buffet')
elif umair_total<zuaina_total:
    print('zuaina wins the buffet')
else:
    print('Error')
```
