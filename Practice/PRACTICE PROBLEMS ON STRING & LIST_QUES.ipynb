### Task 1
Write a python program that will take a number as an input and check whether the number is Awesome or not. If the number is Awesome, it prints True. Otherwise False.

**Awesome number**: a number where every digit is less than its immediate left digit is called an Awesome number. A single-digit number cannot be an awesome number(e.g. 5421 is an Awesome number.)

==========================================================

**Sample Input1:**\
976321\
**Sample Output1:**\
True

==========================================================

**Sample Input2:**\
9766321\
**Sample Output2:**\
False

==========================================================

**Sample Input3:**\
9\
**Sample Output3:**\
False

==========================================================

**Sample Input4:**\
78\
**Sample Output4:**\
False

==========================================================

**Sample Input5:**\
87\
**Sample Output4:**\
True


```python
given = int(input())
given = [int(x) for x in str(given)]
new = []
right = given[0]+1

for i in range(len(given)):
    if right > given[i]:
        new.append(given[i])
        right = given[i]

if len(given) == 1:
    print(False)
elif len(given) == len(new):
    print(True)
else:
    print(False)
```

```python
def check_awesome(n: str) -> bool:
    """
    This will take string and return boolean value
    """
    if len(n) == 1: return False
    for i in range(len(n)-1):
        if n[i] > n[i+1]:
            continue
        else:
            return False
    
    
    return True


print(check_awesome("976321"))
print(check_awesome("9766321"))
print(check_awesome("9"))
print(check_awesome("78"))
print(check_awesome("87"))

print(check_awesome.__doc__)
```
