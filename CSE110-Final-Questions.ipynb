### Question 01
Assume, you have been given a dictionary where the keys present the name of 
the father and the value is a list of names of the sons of that person. For 
example A has 3 sons namely X, Y, and Z. Again X has three sons namely E, 
F, and G. So, A is the grandfather of E, F, and G.

family = {"A" : ["X", "Y", "Z"], "B": ["M", "N"], "W" : ["A", "B"], "X" : 
["E", "F", "G"]}

You need to write a code which takes an input and prints the names of all 
his grandsons in the format shown in output samples. If he does not have 
any grandchildren just print “Get your sons married first! 
Wanna_be_grandpa!!”

You can assume the input will always be a key from the dictionary. But we 
will check your code with a different dictionary. So do not write a code 
for this particular dictionary only.

================================================

Input sample 1:

A

Output sample 1:

E, F, and G

Explanation1:

A has three sons X, Y, and Z but only one of them namely X has sons. So the 
sons of X will be the grandsons of A. So the output is E, F, and G.

================================================

Input sample 2:

W

Output sample 2:

X, Y, Z, M, and N

Explanation2:

Since W has two sons A and B and each of them has sons in return. The sons 
of A namely X, Y, and Z, and the sons of B namely M and N are the grandsons 
of W.

================================================

Input sample 3:

B

Output sample 3:

Get your sons married first! Wanna_be_grandpa!!

Explanation3:

B has sons M and N but none of them have sons. So no grandsons for B.

````python


def family_tree(grandparent):
    family = {"A" : ["X", "Y", "Z"], "B": ["M", "N"], "W" : ["A", "B"], "X" : ["E", "F", "G"]}

    if grandparent in family:
        grandchildren = []
        for parent in family.get(grandparent):
            if family.get(parent):
                grandchildren += list(family.get(parent))
        
        return ", ".join(grandchildren[:-1]) + ' and ' + grandchildren[-1] if grandchildren else "Get your sons married first! Wanna_be_grandpa!!"

print(family_tree("A"))
print(family_tree("W"))
print(family_tree("B"))
````

summation of all the elements in the new list is also a prime number then 
your program should print “It’s a miracle!!!”, otherwise your program 
should print “Just another normal day!!!”. If there is no prime number in 
the given list then keep the new list empty.

======================================================

Sample Input 1:

10, 12, 2, 17, 14


Sample Output 1:

Input list: [10, 12, 2, 17, 14]

Prime list: [2, 17]

It’s a miracle!!!


Explanation 1:

 From the input list, 2 and 17 are prime numbers. So they are added to the

````python

def check_prime(n):
    if n < 2: return False
    if n == 2: return True
    for div in range(2, n//2+1):
        if n % div  == 0: return False
    return True


def miracle_day(string: str):
    given_list = [int(x.strip()) for x in string.split(',')]
    prime_list = [x for x in given_list if check_prime(x)]

    print(f"Input list: {given_list}")
    print(f"Prime list: {prime_list}")

    if check_prime(sum(prime_list)): print("Miracle Day!!!")
    else: print("Just another normal day!!!")
    print()

miracle_day("10, 12, 2, 17, 14")
miracle_day("10, 12, 14")
miracle_day("10, 12, 13, 3")
````

{'Alex': ['Carl', 'Ethan'], 'Bob': ['Carl', 'David'], 'Carl': ['David', 'Alex'], 'David': ['Alex']}

{'Carl': ['Alex', 'Bob'], 'Ethan': ['Alex'], 'David': ['Bob', 'Carl'], 'Alex': ['Carl', 'David']}

### Question 08

Write a python function called replacer()  that takes two strings as 
arguments.
             Argument1: A sentence where you have to make the changes.
             Argument2: A word with the values either “vowel” or “consonant”.
Your function should replace the characters of the sentence with “V” or “C” 
depending on the 2nd argument. If the word is “vowel”, then replace all the 
vowels with “V”. If the word is “consonant”, then replace all the 
consonants with “C”.

[You are not allowed to use the replace() function]

================================================

Function call1:

replacer("It is MAT110 final", "vowels")

Sample Output1:

Vt Vs MVT110 fVnVl


================================================

Function call2:

replacer("It is CSE110 final", "consonants")

Sample Output 02:

IC iC CCE110 CiCaC


================================================

````python
def replacer(sentence: str, category) -> str:
    vowels = "aeiou"
    string = ''

    if "vowels" in category: 
        for char in sentence:
            if char.lower() in vowels: string += 'V'
            else: string += char 

    elif "consonants" in category: 
        for char in sentence:
            if 'a' <= char.lower() <= 'z' and char.lower() not in vowels: string += 'C'
            else: string += char

    return string

print(replacer("It is MAT110 final", "vowels"))
print(replacer("It is CSE110 final", "consonants"))
````

````python

def list_merger(l1, l2):
    l1, l2 = list(eval(l1)), list(eval(l2))
    initial_merge = l1+l2 if len(l1) > len(l2) else None
    if initial_merge:
        updated_merge = initial_merge
        if len(initial_merge) % 2 == 0:
            updated_merge = initial_merge[:len(initial_merge)//2-1] + initial_merge[len(initial_merge)//2+1:]
        else:
            updated_merge = initial_merge[:len(initial_merge)//2] + initial_merge[(len(initial_merge)+1)//2:]
        
        print(f"Initial merged list: {initial_merge}") 
        print(f"Updated merged list: {updated_merge}") 
    else: print("Length of second list is greater than or equal to the length of first list.")
    
# TEST CASES
list_merger("2,3,4,5,6,7,8", "9,10,15,14"), print()
list_merger("2,3,4,8,9,13,15,18", "9,10,15,1")
````
